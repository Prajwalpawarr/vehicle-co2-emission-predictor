```python
model_1 = smf.ols('co2_emissions ~ Q("fuel_consumption_comb(mpg)")', data=df_scaled).fit()
model_2 = smf.ols('co2_emissions ~ transmission', data=df_scaled).fit()
model_3 = smf.ols('co2_emissions ~ fuel_type', data=df_scaled).fit()
model_4 = smf.ols('co2_emissions ~ engine_size', data=df_scaled).fit()
model_5 = smf.ols('co2_emissions ~ engine_size + Q("fuel_consumption_comb(mpg)")', data=df_scaled).fit()
model_6 = smf.ols('co2_emissions ~ engine_size  + Q("fuel_consumption_comb(mpg)") + fuel_type', data=df_scaled).fit()
model_7 = smf.ols('co2_emissions ~ engine_size  + Q("fuel_consumption_comb(mpg)") + transmission', data=df_scaled).fit()
model_8 = smf.ols('co2_emissions ~ engine_size  + Q("fuel_consumption_comb(mpg)") + fuel_type + transmission', data=df_scaled).fit()
```

```python
def model_selector(i):
    switcher = {
        1: model_1,
        2: model_2,
        3: model_3,
        4: model_4,
        5: model_5,
        6: model_6,
        7: model_7,
        8: model_8
    }
    return switcher.get(i, "Invalid option")
```

```python
r_square_values = []
mse=[]
rmse=[]
for i in range(1,9):  # Corrected the range to start from 1 and end at 8
    residuals = model_selector(i).resid
    mse = np.mean(residuals ** 2)
    rmse.append(np.sqrt(mse).round(2))
    r_square_values.append(model_selector(i).rsquared.round(5))

print(rmse)
print(r_square_values)
```

```python
model_names = ['model_1', 'model_2', 'model_3', 'model_4', 'model_5', 'model_6','model_7','model_8']

# Combine model names and R-squared values into a dictionary
data = {'Model': model_names, 'R_squared': r_square_values,'RMSE':rmse}

# Create a DataFrame from the dictionary
df_4 = pd.DataFrame(data)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
```

```python
# Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(SS_X,y,test_size=0.30)
```

# **1] Multilinear Regression**

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

linreg.fit(X_train,Y_train)

Y_pred_train = linreg.predict(X_train)
Y_pred_test  = linreg.predict(X_test)


from sklearn.metrics import mean_squared_error
train_error = np.sqrt(mean_squared_error(Y_train,Y_pred_train))
test_error = np.sqrt(mean_squared_error(Y_test,Y_pred_test))

print("Training Mean sqaure  Error:", train_error.round(2))
print("Test  Mean sqaure Error:", test_error.round(2))
print("variance:",(np.mean(test_error)-np.mean(train_error)).round(2))
```

```python
linreg.score(X_train,Y_train) #Training accuracy is 88.156%
```

```python
linreg.score(X_test,Y_test) #Training accuracy is 88.80%7
```

```python
#The variance is low between the test and train error so the model is consistent both in training and test
```

# **2] Ridge Regression**

Lets find out best value for 'alpha' parameter with the help of Grid Serch CV

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define the Ridge Regression model
ridge=Ridge()

#  range of alpha values to search through
param_grid = {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]}

# Cross Validation
grid_search = GridSearchCV(ridge, param_grid, cv=5)
grid_search.fit(X_train, Y_train)

grid_search.best_params_
```

```python
ridge_reg=Ridge(alpha=1)
ridge_reg.fit(X_train,Y_train)
```

```python
ridge_reg.score(X_train,Y_train)*100 #Training accuracy is 88.15%
```

```python
ridge_reg.score(X_test,Y_test)  #Testing accuracy is 88.80%
```

# **3] Lasso Regression**

Lets find out best value for 'alpha' parameter with the help of Grid Serch CV




```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Define the Lasso Regression model
lasso=Lasso()

#  range of alpha values to search through
param_grid = {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]}

# Cross Validation
grid_search = GridSearchCV(lasso, param_grid, cv=5)
grid_search.fit(X_train, Y_train)

grid_search.best_params_
```

```python
lasso_reg=Lasso(alpha=0.1)
lasso_reg.fit(X_train,Y_train)
print(lasso_reg.score(X_train,Y_train)*100, '% Training accuracy ')
print(lasso_reg.score(X_test,Y_test)*100,  '% Testing accuracy')
```

# **4] Support Vector Machines**

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svm_reg = SVR(kernel='linear')

#  C values to search through
param_grid = {
    'C': [0.1, 1, 10, 100]
}
#  Cross Validation
grid_search = GridSearchCV(svm_reg, param_grid, cv=5)
grid_search.fit(X_train, Y_train)

best_params = grid_search.best_params_
svm_best = SVR(kernel='linear', **best_params)

svm_best.fit(X_train, Y_train)

print("Training Accuracy:", svm_best.score(X_train, Y_train) * 100, "%")
print("Testing Accuracy:", svm_best.score(X_test, Y_test) * 100, "%")
```

# **5] Random Forest Regression**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV


rf_reg = RandomForestRegressor()

#  range of hyperparameters to search
param_dist = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [10, 20, 30, 40, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1],
}

# Randomized Search Cross Validation
random_search = RandomizedSearchCV(rf_reg, param_distributions=param_dist, n_iter=100, cv=5, random_state=42)
random_search.fit(X_train, Y_train)

# best hyperparameters found
best_params = random_search.best_params_

rf_best = RandomForestRegressor(**best_params)

# Fit the model with the best hyperparameters to the training data
rf_best.fit(X_train, Y_train)

print("Training Accuracy:", rf_best.score(X_train, Y_train) * 100, "%")
print("Testing Accuracy:", rf_best.score(X_test, Y_test) * 100, "%")
```

A Random Forest Regression model achieving 99% accuracy on both training and testing datasets is quite unusual and may indicate overfitting

While Random Forest Regression may provide high accuracy in this case, it's crucial to consider other factors such as model interpretability, computational complexity, and deployment requirements. If interpretability is essential or if computational resources are limited, simpler models like Linear Regression or Ridge Regression may still be preferred despite slightly lower accuracy.
